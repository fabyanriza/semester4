````python
list_angka = [8,7,6,5,34,4,3,2,1]

def bubblesort(var):

    iterasi = 0
    while iterasi < (len(var)-1):
    
        cursor = 0
        while cursor < (len(var)-1):

            if var[cursor] > var[cursor + 1]:
                var[cursor], var[cursor + 1] = var[cursor + 1], var[cursor]
            cursor = cursor + 1
        iterasi = iterasi + 1
    return var

print(bubblesort(list_angka))
````

````python
var = [3,5,3,2,6,7,6,4]

def selectionsort(var):
    for i in range(len(var)):

        min_idx = i
        for j in range(i + 1, len(var)):
            if var[min_idx] > var[j]:
                min_idx = j

        var[i], var[min_idx] = var[min_idx], var[i]
    return var

print(selectionsort(var))
````

2. Modifikasilah kode Sorting diatas sehingga menampilkan pengurutan Descending

````python
var = [1,2,3,6,3,2,43,8,3]

def bubblesort(var):
    iterasi = 0
    while iterasi < (len(var)-1):
        cursor = 0
        while cursor < (len(var)-1):
            if var[cursor] > var[cursor + 1]:
                var[cursor], var[cursor + 1] = var[cursor + 1], var[cursor]
            cursor = cursor + 1
        iterasi = iterasi + 1
    var_desc = var[::-1]
    return var_desc

print(bubblesort(var))
````
