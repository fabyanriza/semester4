Buatlah sebuah program pembalik kata. Langkah pertama, tiap katakter pada String input diekstrak dan dimasukkan ke dalam stack. Kemudian tiap karakter tersebut dikeluarkan dan ditampilkan sebagai output. Karena memiliki sifat LIFO, maka keluaran stack adalah karakter-karakter dengan urutan yang berkebalikan dengan input. Program ini memiliki 3 module yaitu<br>a. Stack untuk menyimpan setiap karakter input pada stack array yang berisi constructor dan function operasi stack,<br>b. Pembalik untuk membaca setiap karakter input, menyimpan karakter dengan memanggil function push() pada class “stack” (point a), dan membalik input dengan memanggil function pop() pada class “stack” (point a). Tiap karakter keluaran dari stack tersebut disimpan/ditambahkan (append) pada String output sebagai keluaran yang akan ditampilkan,<br>c. Pembalik Kata ini berisi function main untuk deklarasi dan inisialisasi input, memanggil file Pembalik untuk membalik input dan mendapatkan output kata yang telah dibalik, serta menampilkan output pada console

````python
class Stack:
    def __init__(self):
        self.stack = []
        
    def push(self, char):
        self.stack.append(char)
        
    def pop(self):
        return self.stack.pop()

class reverse:
    def __init__(self, string_input):
        self.string_input = string_input
        self.stack = Stack()
        self.string_output = ""
        
    def reverse_str(self):
        for i in self.string_input:
            self.stack.push(i)
            
        while len(self.stack.stack) != 0:
            self.string_output += self.stack.pop()
            
        return self.string_output

class pembalik_kata:
    def main(self):
        string_input = input("masukkan sebuah kata/string: ")
        pembalik = reverse(string_input)
        string_output = pembalik.reverse_str()
        print(f"string sebelum dibalik: {string_input}")
        print(f"string setelah dibalik: {string_output}")

output = pembalik_kata()
output.main()
````

Buatlah program antrian dengan mengimplementasikan Queue Stack. Adapun aturannya
sebagai berikut:<br>a. penambahan objek pada daftar antrian. Lakukan beberapa kali hingga antrian penuh(maksimum 5). Ketika objek ditambahkan pada antrian yang penuh maka program akan menampilkan keterangan antrian penuh.<br>b. Menampilkan isi antrian <br>c. Satu persatu objek keluar antrian hingga antrian kosong

````python
class Antrian:
    def __init__(self):
        self.antrian = []
        self.maks = 5
        
    def masuk(self, data_antrian):
        for i in data_antrian:
            if len(self.antrian) < self.maks:
                self.antrian.append(i)
                print(f"{i} masuk antrian")
            else:
                print(f"maaf {i}, antrian sudah penuh")
        print(f"data antrian: {self.antrian}")
              
              
    def keluar(self):
        for i in range(len(self.antrian)):
            i_keluar = self.antrian.pop(0)
            self.antrian.append("entek")
            print(f"{i_keluar} keluar antrian")
            print(self.antrian)
            if self.antrian.count("entek") == 5:
                print(f"jumlah antrian : 0 orang")
                break

def main():
    antrian = Antrian()
    data_antrian = []
    while True:
        data = input("Masukkan data (ketik 'selesai' untuk berhenti): ")
        if data.lower() == "selesai":
            break
        data_antrian.append(data)
    antrian.masuk(data_antrian)
    antrian.keluar()

    
main()
````

Buatlah program penelusuran tree dengan menggunakan metode preorder, inorder dan postorder traversal, kemudian tentukan leaf, dan root dari tree berikut<br>  <p style="text-align: center;"></p>

````python
class Node:
    def __init__(self, nilai):
        self.nilai = nilai
        self.kiri = None
        self.kanan = None
        self.tengah = None
        
class PohonBinary:
    def __init__(self,root):
        self.root = Node(root)

    def preorder(self, awal, traversal):
        if awal:
            traversal += (str(awal.nilai) + ' ')
            traversal = self.preorder(awal.kiri, traversal)
            traversal = self.preorder(awal.tengah, traversal)
            traversal = self.preorder(awal.kanan, traversal)
        return traversal
        
    def inorder(self, awal, traversal):
        if awal:
            traversal = self.inorder(awal.kiri, traversal)
            traversal += (str(awal.nilai) + ' ')
            traversal = self.inorder(awal.tengah, traversal)
            traversal = self.inorder(awal.kanan, traversal)
        return traversal
    
    def postorder(self, awal, traversal):
        if awal:
            traversal = self.postorder(awal.kiri, traversal)
            traversal = self.postorder(awal.tengah, traversal)
            traversal = self.postorder(awal.kanan, traversal)
            traversal += (str(awal.nilai) + ' ')
        return traversal
    
    def search_leaf(self, root):
        if root is None:
            return []
        if root.kiri is None and root.tengah is None and root.kanan is None:
            return [root.nilai]
        leaf = []
        if root.kiri is not None:
            leaf += self.search_leaf(root.kiri)
        if root.tengah is not None:
            leaf += self.search_leaf(root.tengah)
        if root.kanan is not None:
            leaf += self.search_leaf(root.kanan)
        return leaf

    
# kasta pertama (root)  
pohon = PohonBinary('Z')
# kasta kedua
pohon.root.kiri = Node('A')
pohon.root.tengah = Node('G')
pohon.root.kanan = Node('R')
# kasta ketiga
pohon.root.kiri.kiri = Node('X')
pohon.root.kanan.kanan = Node('F')
# kasta keempat
pohon.root.kiri.kiri.kiri = Node('B')
pohon.root.kiri.kiri.tengah = Node('N')
pohon.root.kiri.kiri.kanan = Node('W')
pohon.root.kanan.kanan.kiri = Node('S')
pohon.root.kanan.kanan.kanan = Node('J')
# kasta kelima
pohon.root.kiri.kiri.kanan.kiri = Node('Q')
pohon.root.kanan.kanan.kiri.kiri= Node('D')
pohon.root.kanan.kanan.kiri.kanan= Node('K')
pohon.root.kanan.kanan.kanan.kanan= Node('P')

# algoritma

print(pohon.preorder(pohon.root, ""))
print(pohon.inorder(pohon.root, ""))
print(pohon.postorder(pohon.root, ""))

# mencari leaf dan root
print(pohon.search_leaf(pohon.root))
print(pohon.root.nilai)
````
