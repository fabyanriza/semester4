```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Import library yang akan digunakan saat melakukan scraping data

```python
response = requests.get('https://unair.ac.id/news')
rawhtml = response.text
soup = BeautifulSoup(rawhtml, 'html.parser')
```

menggunakan library requests untuk mendapatkan data yang ada pada website yang diambil (UNAIR NEWS), dimana permintaan tersebut di deklarasikan pada variabel response. Dan rawhtml merupakan variabel yang berisikan isi dari website UNAIR NEWS dalam bentuk text.

```python
berita = soup.find_all(class_ = 'elementor-post__title')
df = []
for judul in berita:
    df.append(judul.text)
```

mencari/ atau menarik semua elemen html yang memiliki class 'element-post__title' yang mana elemen tersebut merupakan elemen yang memberikan tanda bahwa teks tersebut merupakan judul berita. Setelah itu meletakkan semua judul kedalam sebuah wadah yang bernamakan df dengan fungsi append

```python
judul_fix = [i.replace('\n\n\t\t\t\t', '').rstrip('\t\t\t\n') for i in df]
judul_fix
```

## NO.2

Crawling unair news dengan mengkombinasikan Python Request, beautifulsoup, dan for loop, untuk mengambil judul berita dari kategori featured news




```python
tempat = []

for page_num in range(1, 3):
    url = f"https://unair.ac.id/category/featured/page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    title = soup.find_all(class_="elementor-post__title")
    
    for tag in title:
        tempat.append(tag.text)

tempat_fix =  [i.replace('\n\n\t\t\t\t', '').rstrip('\t\t\t\n') for i in tempat]
tempat_fix
```

Mengambil data yang memiliki elemen class 'elementor-post__title'. Dan menggunakan metode loop untuk mengambil data atau judul berita di halaman lainnya dengan metode yang sama

```python
isi = pd.DataFrame(tempat_fix)
isi
```

Setelah semua data telah diambil dengan metode crawling, maka hasil / output dari text tersebut akan di jadikan sebuah dataframe / tabel agar lebih mudah untuk dibaca
